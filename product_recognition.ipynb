```python
scenes_files = ['e1.png', 'e2.png', 'e3.png', 'e4.png', 'e5.png']
products_files = ['0.jpg', '1.jpg', '11.jpg', '19.jpg', '24.jpg', '25.jpg', '26.jpg']
```

```python
scenes = [cv2.imread(os.path.join('images/scenes', scene), 0) for scene in scenes_files]
products = [cv2.imread(os.path.join('images/models', product), 0) for product in products_files]
```

```python
plt.imshow(scenes[2], cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
plt.imshow(products[4], cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
sift = cv2.SIFT_create()

kp_scenes = [sift.detect(scene) for scene in scenes]
kp_products = [sift.detect(product) for product in products]
```
